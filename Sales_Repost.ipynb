```python
#loading Data

import pandas as pd
import io

data = """Product,Sales
Laptop,125000
iPhone,98000
T-Shirt,45000
Jeans,32000
Watch,28000
Headphones,67000
Shoes,89000
Sweater,34000
iPad,115000
Mouse,23000
Jacket,52000
Charger,18000
Smart TV,65000
Notebook,12000
Pens,8000"""

df = pd.read_csv(io.StringIO(data))
print(df)
```

```python
# Total Sales
print("Total Sales:",df['Sales'].sum())
```

```python
#Average Sales
print("Average Sales:",df['Sales'].mean())
```

```python
print("Max Sale:", df["Sales"].max())
```

```python
grouped = df.groupby("Product")['Sales'].sum()
print(grouped)
```

```python
#Best Product
print(grouped.idxmax())
```

```python
#clean data

df = df.drop_duplicates()
df["Sales"] = df["Sales"].fillna(df["Sales"].mean())
```

```python
#Sorting values
print(df.sort_values(by="Sales", ascending=False))
```
